```python
from InstructorEmbedding import INSTRUCTOR
model = INSTRUCTOR('hkunlp/instructor-large')
```

## 역대각선 방향의 문장이 positive pair에 해당

```python
from sklearn.metrics.pairwise import cosine_similarity
sentences_a = [['Represent the Science sentence: ','Parton energy loss in QCD matter',0],
               ['Represent the Financial statement: ','The Federal Reserve on Wednesday raised its benchmark interest rate.',0]]
sentences_b = [['Represent the Financial statement: ','The funds rose less than 0.5 per cent on Friday',0],
               ['Represent the Science sentence: ','The Chiral Phase Transition in Dissipative Dynamics', 0]
               ]
embeddings_a = model.encode(sentences_a)
embeddings_b = model.encode(sentences_b)
similarities = cosine_similarity(embeddings_a,embeddings_b)
print(similarities)
```

## 연관 문장끼리 clustering이 잘 되는지 확인

```python
import sklearn.cluster
sentences = [['Represent the Medical sentence for clustering: ','Patient with a history of GERD and Peptic ulcer complains of nausea vomitting', 0],
             ['Represent the Medical sentence for clustering: ','Patient with a history of fatty liver ', 0],
             ['Represent the Medical sentence for clustering: ','Patient complains of severe headache and vision disturbances', 0],
             ['Represent the Medical sentence for clustering: ','Patient complains of swelling of wrist after fall probably fratcure',0],
             ['Represent the Medical sentence for clustering: ','Patient compains of ankle swelling following a slip when walking',0]]
embeddings = model.encode(sentences)
clustering_model = sklearn.cluster.MiniBatchKMeans(n_clusters=3)
clustering_model.fit(embeddings)
cluster_assignment = clustering_model.labels_
print(cluster_assignment)
```
