# 몬테카를로 시뮬레이션으로 배우는 확률 통계 (이산형 분포)

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (3,2)
```

## 이산형 균일 분포

### pmf

```python
def draw_uniform_pmf(a,b):
    x_list = list(range(a, b+1))
    
    p_list = [1/(b-a+1) for x in x_list]
    
    plt.bar(x_list, p_list)
    plt.show()

draw_uniform_pmf(1,5)
```

## 베르누이 분포

### pmf

```python
import matplotlib.pyplot as plt

def draw_bernoulli_pmf(p):
    x_list = [0,1]

    p_list = [(p**x)*(1-p)**(1-x) for x in x_list]
    
    plt.bar(x_list, p_list)
    plt.show()

draw_bernoulli_pmf(p=0.3)
```

## 이항 분포

### pmf

```python
def factorial(x):
    x_list = list(range(1,x+1))
    res = 1
    for val in x_list:
        res *= val
    return res

def combination(n, x):
    res = factorial(n) / (factorial(x)*factorial(n-x))
    return res


def draw_binomial_pmf(n, p):
    x_list = list(range(0, n+1))
    
    p_list = []
    for x in x_list:
        prob = combination(n,x) * (p**x) * ((1-p)**(n-x))
        p_list.append(prob)
    
    plt.bar(x_list, p_list)
    plt.figure(figsize=(3,2))
    plt.show()

draw_binomial_pmf(n=10, p=0.3)
```

### 이항 분포 난수 생성

```python
import numpy as np
x_list = np.random.binomial(n=10, p=0.3, size=10000)
plt.hist(x_list)
plt.show()
```

## 포아송 분포

```python
def draw_poisson_pmf(lamb, n):

    e = 2.7182818284
    x_list = list(range(0,n))
    
    p_list = []
    for x in x_list:
        prob = ((e**(-lamb))*(lamb**x)) / factorial(x)
        p_list.append(prob)
    
    plt.bar(x_list, p_list)
    plt.show()

draw_poisson_pmf(lamb = 2, n = 10)
```

### 포아송 분포 난수 생성

```python
import numpy as np
x_list = np.random.poisson(lam=2.0, size=10000)
plt.hist(x_list)
plt.show()
```

## 기하 분포

```python
def draw_exponential_pmf(p, n):

    x_list = list(range(0,n))
    
    p_list = []
    for x in x_list:
        prob = p*((1-p)**(x-1))
        p_list.append(prob)
    
    plt.bar(x_list, p_list)
    plt.show()

draw_exponential_pmf(p = 0.7, n = 10)
```

### 기하 분포 난수 생성

## 음이항 분포

```python
def draw_negative_binomial_pmf(r, p):

    x_list = list(range(r,r+20))
    
    p_list = []
    for x in x_list:
        prob = combination(x-1,r-1)*(p**r)*((1-p)**(x-r))
        p_list.append(prob)
    
    plt.bar(x_list, p_list)
    plt.show()

draw_negative_binomial_pmf(r=5, p=0.6)
```
