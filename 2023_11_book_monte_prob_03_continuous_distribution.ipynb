# 몬테카를로 시뮬레이션으로 배우는 확률 통계 (연속형 분포)

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (3,2)
```

## 연속형 균일 분포

### pmf

```python
def seq(start, stop, step):
    res = []
    current = start
    while current < stop:
        res.append(current)
        current += step
    return res

def draw_uniform_pdf(a,b):
    x_list = seq(a,b,0.01)
    
    p_list = [1/(b-a) for x in x_list]
    
    plt.plot(x_list, p_list)
    plt.show()

draw_uniform_pdf(1,5)
```

### 연속형 균일 난수 생성

```python
import numpy as np
x_list = np.random.uniform(low=1.0, high=5.0, size=10000)
plt.hist(x_list)
plt.show()
```

## 정규 분포

```python
def draw_normal_pdf(mu, s):
    pi = 3.1415926535
    e = 2.7182818284
    
    x_list = seq(mu-3*s, mu+3*s, 0.01)
    
    p_list = [ (1/(((2*pi)**0.5)*s))* (e**(-0.5*((x-mu)/s)**2)) for x in x_list]
    
    plt.plot(x_list, p_list)
    plt.show()

draw_normal_pdf(1,5)
```

### 정규 분포 난수 생성

```python
import numpy as np
x_list = np.random.normal(loc=0, scale=3, size=10000)
plt.hist(x_list)
plt.show()
```

## 감마 분포

```python
def gamma_function(alpha, x):
    e = 2.7182818284
    res = (x**(alpha-1)*(e**(-x)))
    return res

def gamma(alpha):
    a = 0
    b = 100
    x_list = seq(a+0.0001, b, 0.001)
    
    gamma_list = []
    for x in x_list:
        y = gamma_function(alpha, x)
        gamma_list.append(y)
    res = ((b-a)/len(x_list)*sum(gamma_list))
    return res

def draw_gamma_pdf(alpha, beta):
    e = 2.7182818284
    
    x_list = seq(0, alpha*beta*5, 0.01)
    
    p_list = []
    for x in x_list:
        prob = (1/(gamma(alpha)*(beta**alpha)))*(x**(alpha-1))*(e**((-1/beta)*x))
        p_list.append(prob)
        
    plt.plot(x_list, p_list)
    plt.show()    
    
draw_gamma_pdf(alpha=2, beta=1)
```

### 감마 분포 난수 생성

## 지수 분포

```python
def draw_exponential_pdf(beta):
    e = 2.7182818284
    
    x_list = seq(0, beta*10, 0.01)
    
    p_list = []
    for x in x_list:
        prob = (1/beta)*(e**((-1/beta)*x))
        p_list.append(prob)
        
    plt.plot(x_list, p_list)
    plt.show()    
    
draw_exponential_pdf(beta=2)
```

## 카이제곱 분포

```python
def draw_chi2_pdf(p):
    e = 2.7182818284

    alpha = p/2
    beta = 2
    
    x_list = seq(0, p*5, 0.01)
    
    p_list = []
    for x in x_list:
        prob = (1/(gamma(alpha)*(beta**alpha)))*(x**(alpha-1))*(e**((-1/beta)*x))
        p_list.append(prob)
        
    plt.plot(x_list, p_list)
    plt.show()    
    
draw_chi2_pdf(p=3)
```

## 베타 분포

```python
def draw_beta_pdf(alpha, beta):
    x_list = seq(0.01, 1, 0.01)
    
    p_list = []
    beta_function = ((gamma(alpha+beta))/(gamma(alpha)*gamma(beta)))
    for x in x_list:
        prob = beta_function*(x**(alpha-1))*((1-x)**(beta-1))
        p_list.append(prob)
        
    plt.plot(x_list, p_list)
    plt.show()    
    
draw_beta_pdf(alpha=0.5, beta=0.5)
```
